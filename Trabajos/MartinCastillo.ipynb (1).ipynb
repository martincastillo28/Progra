###El Pais escogido fue Irlanda y se dividio en 3 zonas:
1.Norte: Corresponde a condados como Donegal y zonas del Ulster.  
2.Centro: Abarca el área de Dublín y sus alrededores.  
3.Sur: Incluye condados como Cork, Limerick y Waterford.

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mticker
class PoblacionIrlanda:
    def __init__(self, regiones, anos, poblaciones):
        self.regiones = regiones
        self.anos =anos
        self.poblaciones = poblaciones

    def graficar(self):
      colores = ['green', 'orange', 'deepskyblue']
      plt.stackplot(self.anos, self.poblaciones, labels=self.regiones, colors=colores)
      plt.title('Población de Irlanda')
      plt.xlabel('Años')
      plt.ylabel('Población (millones)')
      plt.grid(axis='y', linestyle='--', alpha=0.7)
      plt.xticks(self.anos)

      plt.legend(title="Regiones", loc="upper left", facecolor="white",framealpha=1)
      plt.show()

regiones = ['Norte', 'Centro', 'Sur']
anos = [2020, 2021, 2022, 2023, 2024]
poblaciones = [
    [0.55, 0.57, 0.58, 0.60, 0.62],
    [1.90, 1.95, 1.97, 2.00, 2.05],
    [2.10, 2.15, 2.18, 2.20, 2.25]
]

grafico_irlanda = PoblacionIrlanda(regiones, anos, poblaciones)
grafico_irlanda.graficar()

```
